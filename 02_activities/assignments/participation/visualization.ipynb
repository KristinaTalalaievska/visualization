```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python

# Seed the random number generator for reproducibility
np.random.seed(613)

# Create a sequence of 50 values from 0 to 49
x = np.arange(50)

# Generate 50 random integers between 0 and 100
y = np.random.randint(0, 100, 50)

# Make a basic scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)

# Edit axis object and create a bar plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)

# Make a line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

# Make a histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)

# Add the title and labels to the last plot
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')

# Adjust layout to ensure everything fits without overlapping
fig.tight_layout()

# Define font dictionaries for custom styling
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

# Add fontdict arguments to the line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)

# Adjust layout to ensure everything fits without overlapping
fig.tight_layout()

# Show the plots (optional, especially if using a script)
plt.show()

# Remake basic scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(
    x,
    y,
    marker='*',        # Star marker
    color='indigo'     # Indigo color
)
# Show the figure
plt.show()

# Adjust the style and width of the line
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(
    x,
    y,
    marker='*',        # Star marker
    color='indigo',    # Indigo color
    linestyle='--',    # Dashed line style
    linewidth=2        # Line width of 2
)
# Show the figure
plt.show()

#modify data points using colours:
##fig, ax = plt.subplots(figsize=(5,3))
#color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
##try to plot both y variables on the same axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1)
ax.plot(x, y2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc='lower right')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1,
                               figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1,
                               figsize=(9,3))
font1 = {'family': 'serif', 'size':14}
ax1.scatter(x1, y1,
            marker='o',
            color = 'red')
ax2.bar(x2,y2, color = 'darkgreen')
ax1.set_title('Scatterplot!', fontdict=font1)
ax2.set_title('Barplot!', fontdict=font1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize = (7,4))
someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1, y1) 
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print (tips)
```

```python
sns.lineplot(data=tips,
             x='total_bill',
             y='tip')
```

```python
sns.set_style('whitegrid')
```

```python
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip')
tipgraph.set(title='Tips vs.Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10,3))
```

```python
tipgraph = sns.lineplot (data=tips,
                         x='total_bill',
                         y='tip',
                         color ='hotpink',
                         linestyle = '--',
                         linewidth = 3,
                         marker = 'o',
                         markerfacecolor = 'indigo')
```

```python
tipgraph = sns.scatterplot(data=tips, x='total_bill',
                           y='tip', style = 'time', hue =
                           'day', palette = ['purple',
                             'hotpink', 'deepskyblue',
                             'yellowgreen'])
tipgraph.set(title='Tips vs Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
sns.pairplot(
    data = tips,
    hue = 'day')
```

```python
daysplot = sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    col="day",
    kind="scatter",
    col_wrap=2)
```

```python

import pandas as pd
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", 
                 on_bad_lines='skip')

print(df.head())
```

```python
text = "".join(each for each in df.quote)
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)
fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,
          interpolation='bilinear')
ax.axis("off")
plt.show()
```

```python

from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

diagram = venn2([A, B],
                set_labels=("Fruits", "Pies"),
                set_colors=("blue", "red"),
                alpha = 0.5
                )

diagram.get_label_by_id("10").set_text("\n".join(A-B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B-A))
```
